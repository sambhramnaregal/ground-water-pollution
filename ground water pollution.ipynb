```python
df9= pd.read_csv(r"statewisemean1.csv")
```

```python
df9['qnph']=100*((df9['maxpH']-7)/(8.5-7))
```

```python
df9['qnnitrate']=100*((df9['maxNitrate-N + Nitrite-N (mg/L)']-0)/(10-0))
```

```python
df9['qnbod']=100*((df9['maxBOD (mg/L)']-0)/(5-0))
```

```python
df9['qnec']=100*((df9['maxConductivity (æmhos/cm)']-0)/(400-0))
```

```python
df9['qntds']=100*((df9['maxTotal Dissolved Solids (mg/L)']-0)/(500-0))
```

```python
df9['qnfc']=100*((df9['maxFecal Coliform (MPN/100ml)']-0)/(500-0))
```

```python
df9['qntec']=100*((df9['maxTotal Coliform (MPN/100ml)']-0)/(100-0))
```

```python
df9['prod_ph']=df9['qnph']*0.270984351
df9['prod_bod']=df9['qnbod']*0.460673396
df9['prod_nitrate']=df9['qnnitrate']*0.230336698
df9['prod_totalcoli']=df9['qntec']*0.02303367
df9['prod_ec']=df9['qnec']*0.005758417
df9['prod_fc']=df9['qnfc']*0.004606734
df9['prod_tds']=df9['qnnitrate']*0.004606734
```

```python
df9['sumqnwn']=df9['prod_ph']+df9['prod_bod']+df9['prod_nitrate']+df9['prod_totalcoli']+df9['prod_ec']+df9['prod_fc']+df9['prod_tds']
```

```python
sumwn=0.270984351+0.230336698+0.460673396+0.005758417+0.004606734+0.004606734+0.02303367


sumwn
```

```python
df9['WQI']=df9['sumqnwn']/1
df9
```
